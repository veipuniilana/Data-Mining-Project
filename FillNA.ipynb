```python
import pandas as pd
import numpy as np
from sklearn import linear_model
%matplotlib inline
```

```python
def random_imputation(df, feature):

    number_missing = df[feature].isnull().sum()
    observed_values = df.loc[df[feature].notnull(), feature]
    df.loc[df[feature].isnull(), feature] = np.random.choice(observed_values, number_missing, replace = True)
    
    return df

```

```python
places =['andhra_pradesh','assam','bihar','chattissgarh','daman_diu_dadra_nagar_haveli','goa','himachal_pradesh','kerala','lakshadweep','madhya_pradesh','maharashtra','odisha','pondicherry','punjab','rajasthan','tripura','uttar_pradesh_uttarakhand','west_bengal']
```

```python
df1 = pd.DataFrame()


for place in places:
    
    df = pd.read_csv("Ground_Water_Data/ground_water_quality_in_"+place+"-2014.csv",encoding = "unicode_escape")
    
    
    df1 = df1.append(df, ignore_index=True)
    if df1.empty:
        df1 = df
df.head()
```

```python
missing_columns = df1.loc[:, df1.isnull().any()].columns.values
for feature in missing_columns:
    
    df1 = random_imputation(df1, feature)

for feature in missing_columns:
    df1 = random_imputation(df1,feature)
    

df1.to_csv('Ground_water_pre_processed/filled_groundwater.csv')
```
